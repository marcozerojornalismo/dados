# Proposition 63 Analysis

Analysis of proposition 63 - Firearms, ammunition sales and intiative statue.

```python
import pandas as pd
```

```python
props = pd.read_csv('http://www.firstpythonnotebook.org/_static/committees.csv')
```

```python
props.prop_name.value_counts()
```

```python
props.calaccess_committee_id.head()
```

```python
contribs = pd.read_csv('http://www.firstpythonnotebook.org/_static/contributions.csv')
```

```python
prop63 = props [props.prop_name == 'PROPOSITION 063- FIREARMS. AMMUNITION SALES. INTIATIVE STATUTE.']
```

```python
prop64.info()
```

```python
prop64.prop_name.value_counts()
```

```python
contribs.info()
```

```python
#Merge
merged = pd.merge(prop64, contribs, on='calaccess_committee_id')
```

```python
mergedSup = merged [merged.committee_position == 'SUPPORT']
mergedOpp = merged [merged.committee_position == 'OPPOSE']
mergedDif = float(mergedSup.amount.sum()) - float(mergedOpp.amount.sum())
#mergedDif
mergedMedSup = float(mergedSup.amount.sum()) / float(mergedSup.committee_position.value_counts())
#mergedMedSup
mergedMedOpp = float(mergedOpp.amount.sum()) / float(mergedOpp.committee_position.value_counts())
#mergedMedOpp
mergedAvrg = float(mergedMedSup + mergedMedOpp) / 2
mergedAvrg
#mergedtest = mergedSup.committee_position.value_counts()
#mergedtest
```
